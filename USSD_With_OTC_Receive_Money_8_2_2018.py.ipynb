## Now i'll find no of weeks in a month .... and by which i can analysis on which week of the month have average revenue , peak and below average revenue.

## Here below generate a function of finding no of weeks in a month.A column create nam 'wom' to show that a date have in which week of the Month, Then sum all the weeks of the month and show a trend of which week have highest trend and whichone have low trend in the week of months. 

```python
import datetime
import calendar
import pandas as pd

def week_of_month(tgtdate):
    tgtdate = pd.to_datetime(tgtdate)

    days_this_month = calendar.mdays[tgtdate.month]
    for i in range(1, days_this_month):
        d = datetime.datetime(tgtdate.year, tgtdate.month, i)
        if d.day - d.weekday() > 0:
            startdate = d
            break
    # now we canuse the modulo 7 appraoch
    return (tgtdate - startdate).days //7 + 1

```

```python
#len(df.new_date)
df['wom']=df.new_date.apply(week_of_month)
```
