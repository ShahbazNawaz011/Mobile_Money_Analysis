## Read Files

```python
df = pd.read_csv('Complete_Months_Data.csv' , encoding='latin1' , index_col=0)
```

```python
df.shape , df.columns
```

```python
df = (df[(df.CHANNEL == 'USSD') & (df.TRANSACTION_TYPE== 'OTC Bill Payment')])
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
df1 = df[['Month', 'date','Hours','Revnue']]
```

```python
df1 = pd.DataFrame({'sum_inflow' : df1.groupby( ["Month", "date","Hours"] )["Revnue"].sum()}).reset_index()
```

```python
#df.groupby(["Month", "date","Hours"])["Revnue"].transform('count')
```

```python
df1.drop(columns='Fraud')
```

```python
df1.describe().astype(int)
```

```python
Avg_inflow = df1.sum_inflow.mean()
```

```python
df1['Avg_inflow - Sum_inflow'] = df1.sum_inflow - Avg_inflow
```

```python
df1
```

```python
conditions = [
    (df1['Avg_inflow - Sum_inflow'].values >= 0),
    (df1['Avg_inflow - Sum_inflow'].values < 0)
    ]
choices = [1 , 0 ]
df1['Earning1'] = np.select(conditions, choices, default=1)                         
```

## Divide data based on testing and training. 

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3, random_state=0)
```

```python
from collections import Counter
Counter(y_train)
```

```python
from sklearn import svm

clf = svm.SVC(kernel='linear', C=100).fit(x_train, y_train)
clf.score(x_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
clf = svm.SVC(kernel='linear', C=1)
scores = cross_val_score(clf,x_test, y_test, cv=5)
scores 
```

```python
from sklearn import preprocessing
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=0)
scaler = preprocessing.StandardScaler().fit(x_train)
x_train_transformed = scaler.transform(x_train)
clf = svm.SVC(C=1).fit(x_train_transformed, y_train)
x_test_transformed = scaler.transform(x_test)
clf.score(x_test_transformed, y_test)  
```

```python
#Support vector regression
clf = svm.SVC()
#training
clf.fit(x_train,y_train)  
```

```python
#prediction
prd=clf.predict(x_test)
```

```python
#accuracy
clf.score(x_test, y_test)
```

```python
#F1_score
from sklearn.metrics import f1_score
f1_score(y_test, prd, average='weighted')
```

```python
# svm classification
clf = svm.SVC(kernel='rbf', gamma=0.7, C = 1.0).fit(x_train, y_train)
y_predicted = clf.predict(x_test)
```

```python
f1_score(y_test, y_predicted, average='weighted')
```

```python
from sklearn import metrics
# performance
print(metrics.accuracy_score(y_test, y_predicted))
print(metrics.precision_recall_fscore_support(y_test, y_predicted ))
print ("Confusion matrix")
print (metrics.confusion_matrix(y_test, y_predicted))
```
