```python
Total_Deduction = df.groupby('date')['Total_deduction'].sum().astype(int)
```

```python
data = [go.Bar(
            x=Total_Deduction.index,
            y=Total_Deduction.values,
            text=Total_Deduction.values,
            textposition = 'auto',
            marker=dict(color='#BFAFB2 '),

            
    )]
layout = Layout(title="6Months DateWise Deduction Amount",
                xaxis=dict(title='Dates'),
                yaxis=dict(title='AMOUNT (in MILLION)'))

fig = Figure(data=data , layout = layout)
py.iplot(fig)
```

```python
Revenue_Generated = df.groupby('date')['Revnue'].sum()
```

```python
data = [go.Bar(
            x=Revenue_Generated.index,
            y=Revenue_Generated.values,
            text=Revenue_Generated.values,
            textposition = 'auto',
             marker=dict(color='#C39953'),
)]
layout = Layout(title="6Months DateWise Revenue Amount",
                xaxis=dict(title='Dates'),
                yaxis=dict(title='AMOUNT (in MILLION)'))
fig = Figure(data=data, layout=layout)
py.iplot(fig)
```

```python
Revenue_profit = Bar(
            x=Revenue_Generated.index,
            y=Revenue_Generated.values,
            name='Revenue',
            marker=dict(color='#4B3621'),
            text=Revenue_Generated.values,
            textposition = 'auto'
    )

Deduction_Taxes = Bar(
            x=Total_Deduction.index,
            y=Total_Deduction.values,
            name='Deduction',
            marker=dict(color='#A2D5F2'),
            text=Total_Deduction.values,
            textposition = 'auto'
    )

data = ([Revnue_profit, Deduction_Taxes])
layout = Layout(title="6Months Revnue and Deduction Amount",
                xaxis=dict(title='Dates'),
                yaxis=dict(title='AMOUNT (in MILLION)'))
fig = Figure(data=data, layout=layout)

py.iplot(fig)
```

```python
Total_Revnue_Deduction = df.groupby('date')['AMOUNT'].sum()
```

```python
Revnue = Bar(
            x=Revnue_Generated.index,
            y=Revnue_Generated.values,
            name='Revnue',
            text=Revnue_Generated.values,
            textposition = 'auto',
            marker = dict(color='#D2691E')
    )

Deduction = Bar(
            x=Total_Deduction.index,
            y=Total_Deduction.values,
            name='Deduction',
            text=Total_Deduction.values,
            textposition = 'auto',
             marker=dict(color='#68A0B0')
    
    )

data = [Revnue, Deduction]

layout = Layout(title="Total Revnue and Deduction",
                xaxis=dict(title='Dates'),
                yaxis=dict(title='Salary (in thousands)'),
               
                barmode='stack'  )


fig = Figure(data=data, layout=layout)
py.iplot(fig)
```
